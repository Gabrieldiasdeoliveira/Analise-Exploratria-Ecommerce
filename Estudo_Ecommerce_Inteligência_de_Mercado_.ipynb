# Preço médio por Categoria

```r
# Calcule o preço médio por categoria
tabela_preco_medio <- aggregate(PrecoTotal ~ NomeCategoria, data = dados,
FUN = mean)

# Arredonde o preço médio para duas casas decimais
tabela_preco_medio$PrecoTotal <- round(tabela_preco_medio$PrecoTotal, 2)

# Renomeie a coluna de preço médio
colnames(tabela_preco_medio)[2] <- "preco_medio"

# Exiba a tabela com o preço médio por categoria
print(tabela_preco_medio)
```

# Nessa etapa, agrupapos as categorias para entender os status de cada uma delas.

```r
# Agrupar os dados por NomeCategoria e Status e calcular a contagem
CategoriaAgrupada <- aggregate(Quantidade ~ NomeCategoria + Status, data = dados,
FUN = length)

# Reformatar a tabela para ter as categorias como colunas
tabela_final <- reshape(CategoriaAgrupada, idvar = "NomeCategoria",
timevar = "Status", direction = "wide")


# Preencher possíveis valores NA com 0
tabela_final[is.na(tabela_final)] <- 0

# Calcular a soma de cada linha
soma_linhas <- apply(tabela_final[, -1], 1, sum)

# Adicionar coluna de soma à tabela
tabela_final$Soma <- soma_linhas

# Exibir a tabela final
tabela_final
```

```r
# Agregar os dados por NomeCategoria, Status e calcular a soma do PrecoTotal
dados_agrupados2 <- aggregate(PrecoTotal ~ NomeCategoria + Status, data = dados, FUN = sum)

# Aplicar a função pivot_wider() para reformatar a tabela
tabela_final2 <- pivot_wider(dados_agrupados2, names_from = Status, values_from = PrecoTotal, values_fill = 0)

# Adicionar coluna com a soma de cada linha
tabela_final2 <- tabela_final2 %>% mutate(Soma = rowSums(.[-1]))

# Calcular a soma das colunas
soma_colunas <- tabela_final2 %>% summarise(across(-1, sum))

# Atribuir o nome "Total" à última linha
soma_colunas <- soma_colunas %>% mutate(NomeCategoria = "Total")

# Adicionar a linha de soma na tabela
tabela_final2 <- bind_rows(tabela_final2, soma_colunas)

# Exibir a tabela final atualizada
tabela_final2
```

```r
# Extrair o ano da coluna DataCriacao
anos <- unique(substr(dados$DataCriacao, 1, 4))

# Exibir os anos únicos
print(anos)
```

```r
# Extrair o ano da coluna de datas
dados$Ano <- format(dados$Data, "%Y")

# Agrupar o status pelos anos e calcular a soma
soma_status <- aggregate(Quantidade ~ Status + Ano, data = dados, FUN = sum)

# Exibir a tabela resultante
soma_status
```

```r
# Criar o gráfico de barras
grafico <- ggplot(data = soma_status, aes(x = Status, y = Quantidade, fill = as.factor(Ano))) +
  geom_bar(stat = "identity", position = "dodge") +
  labs(x = "Status", y = "Quantidade", fill = "Ano") +
  scale_fill_manual(values = c("#F8766D", "#00BFC4"), labels = c("2019", "2020")) +
  theme_minimal()

# Exibir o gráfico
print(grafico)
```
