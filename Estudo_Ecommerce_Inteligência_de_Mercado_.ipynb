 # Estudo - Ecommerce Inteligência de Mercado

*O uso de inteligência de mercado em uma base de dados como essa pode fornecer insights valiosos e ajudar a tomar decisões com base em dados.
Irei realizar alguns estudos de inteligência de mercado usando uma base de dados de compras.*

*A primeira etapa foi extrair as informações para montar a base de dados.
Foram disponibilizados 5 tabelas.*

*Aqui está uma descrição de cada uma das tabelas:*

* *Tabela "Categories": Colunas: id, name
Essa tabela armazena as categorias dos produtos disponíveis em seu e-commerce. Cada categoria é identificada por um ID exclusivo e possui um nome descritivo.*

* *Tabela "Customers": Colunas: id, created_at, first_name, last_name, email, cell_phone, country, state, street, number, additionals
Essa tabela armazena informações sobre os clientes do seu e-commerce. Cada cliente é identificado por um ID exclusivo e possui detalhes como data de criação, nome, sobrenome, email, número de telefone, país, estado, rua, número e informações adicionais.*

* *Tabela "Items": Colunas: id, order_id, product_id, quantity, total_price
Esta tabela é usada para registrar os itens presentes em cada pedido. Cada item possui um ID exclusivo e está associado a um pedido específico (por meio do "order_id") e a um produto específico (por meio do "product_id"). Também são armazenadas informações sobre a quantidade e o preço total do item.*

* *Tabela "Orders":Colunas: id, created_at, customer_id, status
Nesta tabela, são registrados os pedidos feitos pelos clientes. Cada pedido possui um ID exclusivo e está vinculado a um cliente específico (por meio do "customer_id"). Além disso, são armazenadas informações sobre a data de criação do pedido e o status atual do mesmo.*

* *Tabela "Products":Colunas: id, name, price, category_id
Esta tabela contém informações sobre os produtos disponíveis no seu e-commerce. Cada produto possui um ID exclusivo, um nome, um preço e é associado a uma categoria específica (por meio do "category_id").*
